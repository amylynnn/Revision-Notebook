```python
# Loads the tables from the Wikipedia page

# Identifies the likely population table

# Renames columns safely

# Cleans the population column by removing commas and converting to numeric

# Prints the cleaned DataFrame preview

import pandas as pd

# Step 1: Load all tables from the Wikipedia page
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)'
tables = pd.read_html(url)
print(f"Number of tables found: {len(tables)}")

# Step 2: Choose the table likely containing population data (usually index 1)
population_table = tables[1]

# Step 3: Inspect the table columns and shape
print("Original columns:", population_table.columns.tolist())
print("Shape:", population_table.shape)
print(population_table.head())

# Step 4: Make a copy to work on
df = population_table.copy()

# Step 5: Rename columns if count matches expected
expected_cols = ['Rank', 'Country', 'UN Population', 'Percentage of World Population', 'Date', 'Source']
if len(df.columns) == len(expected_cols):
    df.columns = expected_cols
    print("Columns renamed.")
else:
    print(f"Column count {len(df.columns)} does not match expected {len(expected_cols)}. Skipping renaming.")

# Step 6: Clean the 'UN Population' column: remove commas and convert to float
pop_col = 'UN Population'
if pop_col in df.columns:
    df[pop_col] = (
        df[pop_col]
        .astype(str)
        .str.replace(',', '', regex=False)
        .str.extract('(\d+)')  # Extract digits only
        .astype(float)
    )
    print(f"Cleaned '{pop_col}' column.")
else:
    print(f"Column '{pop_col}' not found.")

# Step 7: Display cleaned data preview
print(df.head())
```
