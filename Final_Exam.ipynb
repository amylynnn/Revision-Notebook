```python
# Manual Rewrite Notes:

tasks = []

def add_task(task):
    """
    Add a new task to the global tasks list.

    >>> tasks.clear()
    >>> add_task('Buy milk')
    Added: Buy milk
    >>> tasks
    ['Buy milk']
    """
    tasks.append(task)
    print(f"Added: {task}")

def remove_task(index):
    """
    Remove a task by its index (0-based). Prints error if index is invalid.

    >>> tasks.clear()
    >>> tasks.extend(['Task1', 'Task2'])
    >>> remove_task(1)
    Removed: Task2
    >>> remove_task(5)
    Invalid index!
    >>> tasks
    ['Task1']
    """
    # Ensure the index is within the valid range
    if 0 <= index < len(tasks):
        removed = tasks[index]
        del tasks[index]
        print(f"Removed: {removed}")
    else:
        print("Invalid index!")

def list_tasks():
    """
    Print all tasks in the tasks list, numbered from 1. If no tasks, print a message.

    >>> tasks.clear()
    >>> list_tasks()
    No tasks available.
    >>> tasks.extend(['Task A', 'Task B'])
    >>> list_tasks()
    1: Task A
    2: Task B
    """
    if not tasks:
        print("No tasks available.")
    else:
        for i, t in enumerate(tasks):
            print(f"{i+1}: {t}")

def main():
    """
    Demonstrate adding, listing, and removing tasks.
    """
    add_task("Buy milk")
    add_task("Pay bills")
    add_task("Walk dog")
    list_tasks()
    remove_task(1)
    list_tasks()

if __name__ == "__main__":
    import doctest
    doctest.testmod()
    main()


"""I manually rewrote the script in my Colab notebook, ensuring:

- All typos were fixed.

- Correct variables are used consistently.

- Function names are correct.

- Added docstring comments with doctest examples to demonstrate testing.

- Comments added to clarify the fixes."""

# Course References:
'''My error handling approach comes from Week 6: Defensive Programming, which emphasized validating inputs and avoiding runtime errors.

My testing approach follows Week 12: Testing and Debugging, where doctests and assert statements were introduced to verify function behaviour.'''
```
