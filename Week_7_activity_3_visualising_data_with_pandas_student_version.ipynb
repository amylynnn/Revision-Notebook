```python
# Activity 1: Line Chart for Monthly Sales

import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Month': ['January', 'February', 'March', 'April', 'May'],
    'Total Sales': [450, 510, 580, 610, 590]
}

df = pd.DataFrame(data)

# Create line chart
ax = df.plot(x='Month', y='Total Sales', kind='line', title='Monthly Sales')

# Add axis labels
ax.set_xlabel('Month')
ax.set_ylabel('Total Sales')

plt.show()
```

```python
# Activity 2: Bar Chart Comparing Products

product_data = {
    'Product': ['A', 'B', 'C', 'D'],
    'Units Sold': [1200, 900, 1350, 1100]
}

product_df = pd.DataFrame(product_data)

# Create bar chart
ax = product_df.plot(x='Product', y='Units Sold', kind='bar', legend=False, title='Units Sold by Product')

# Add axis labels
ax.set_xlabel('Product')
ax.set_ylabel('Units Sold')

plt.show()
```

```python
# Optional Extension: Grouped Bar Chart Example

grouped_data = {
    'Month': ['January', 'January', 'February', 'February', 'March', 'March'],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Sales': [120, 100, 135, 115, 145, 120]
}

grouped_df = pd.DataFrame(grouped_data)

# Pivot so months are rows and products are columns
pivot_df = grouped_df.pivot(index='Month', columns='Product', values='Sales')

# Plot grouped bar chart
pivot_df.plot(kind='bar', title='Monthly Sales by Product')

plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```
