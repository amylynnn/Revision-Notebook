```python
# Activity 1: Writing Your First doctest

import doctest

def add(a, b):
    """
    Adds two numbers and returns the result.

    Examples:
    >>> add(2, 3)
    5
    >>> add(-1, 1)
    0
    >>> add(10, -5)
    5
    """
    return a + b

if __name__ == '__main__':
    doctest.testmod(verbose=True)
```

```python
# Activity 2: Find & Fix a Bug with doctest

import doctest

def multiply_buggy(a, b):
    """
    Supposed to multiply two numbers.

    Examples:
    >>> multiply_buggy(3, 5)
    15
    >>> multiply_buggy(-2, 4)
    -8
    """
    return a + b  # Bug: Should be multiplication

def multiply(a, b):
    """
    Multiplies two numbers correctly.

    Examples:
    >>> multiply(3, 5)
    15
    >>> multiply(2, 0)
    0
    >>> multiply(-2, 4)
    -8
    >>> multiply(0.5, 4)
    2.0
    """
    return a * b

if __name__ == '__main__':
    print("Testing corrected multiply function:")
    doctest.testmod(verbose=True)
```

```python
# Extension Challenge: Subtraction with doctest

import doctest

def subtract(a, b):
    """
    Subtracts b from a.

    Examples:
    >>> subtract(10, 3)
    7
    >>> subtract(0, 5)
    -5
    >>> subtract(-2, -3)
    1
    >>> subtract(5.5, 2.2)
    3.3
    >>> subtract(3, 3)
    0
    """
    return a - b

if __name__ == '__main__':
    doctest.testmod(verbose=True)
```
